#Data Preprocessing and Data Cleaning

```python
# Showing Data Types of the columns
terrorism.dtypes
```

```python
# Extracting Important Columns
terrorism = terrorism[['iyear','imonth','iday','country_txt','provstate','region_txt',
                       'attacktype1_txt','target1','nkill','nwound','summary','gname',
                       'targtype1_txt','weaptype1_txt','motive']]
```

```python
# Checking for columns
terrorism.columns
```

```python
# Renaming of Columns
terrorism.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'State',
                       'region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target','nkill':'Killed',
                       'nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type',
                       'weaptype1_txt':'Weapon_type','motive':'Motive'},inplace=True)
```

```python
terrorism.head(2)
```

```python
# Checking for duplicated values
terrorism.duplicated().sum()
```

```python
# Dropping Duplicated Values
terrorism.drop_duplicates(inplace=True)
terrorism.shape
```

```python
# Checking for duplicated values
terrorism.duplicated().any()
```

```python
# Checking No. of unique values in the columns
terrorism.nunique()
```

#Null Values Treatment

```python
# Checking for null values
terrorism.isna().sum().sort_values()
```

```python
# Counting Values of State Column
terrorism['State'].value_counts()
```

```python
# Treating Null Values in State Column
terrorism['State'] = terrorism['State'].replace({'Unknown': np.nan})
terrorism['State'] = terrorism['State'].fillna("Others")
terrorism['State'].value_counts()
```

```python
# Checking for null values
terrorism['State'].isna().any()
```

```python
# Counting values for Target Column
terrorism['Target'].value_counts()
```

```python
# Treating Null Values in Target Column
terrorism['Target'] = terrorism['Target'].replace({'Unknown': np.nan})
terrorism['Target'] = terrorism['Target'].fillna("Others")
terrorism['Target'].value_counts()
```

```python
# Checking for null values
terrorism['Target'].isnull().any()
```

```python
# Counting Values for Killed Column and Wounded Column
print(terrorism['Killed'].value_counts())
print(terrorism['Wounded'].value_counts())
```

```python
# Checking for Outliers in Killed and Wouded Column
terrorism[['Killed','Wounded']].plot(kind='box')
```

```python
# Treating Null Values in Killed and Wounded Column with median
terrorism[['Killed', 'Wounded']] = terrorism[['Killed', 'Wounded']].fillna(terrorism[['Killed', 'Wounded']].mean())
terrorism[['Killed','Wounded']].isnull().any()
```

```python
# Dropping Null Values of Summary and Motive Column
terrorism = terrorism.dropna(subset=['Summary'])
terrorism = terrorism.dropna(subset=['Motive'])
```

```python
# Checking for null values
terrorism[['Summary','Motive']].isna().any()
```

```python
# checking Unique Values of Day Column
terrorism['Day'].unique()
```

```python
# Changing Data of Day Column
terrorism['Day'] = terrorism['Day'].replace(0,np.nan)
terrorism = terrorism.dropna(subset=['Day'])

# Checking no. of unique values in Day Column
print(terrorism['Day'].nunique())

# Checking for null values
print(terrorism['Day'].isnull().any())
```

```python
# Checking for Null Values
terrorism.isnull().sum()
```

#1. Year Column


1.   Attacks Per Year




```python
# Plotting Countplot to show Attacks per Year
plt.figure(figsize=(15,5))
sns.countplot(x = terrorism['Year'])
plt.title('Terrorist Activities per Year')
plt.xlabel('Attacking Years')
plt.ylabel('No. of Attacks per Years')
plt.xticks(rotation=90)
plt.show()
```

Insights

  1. Over time, there has been a noticeable rise in the frequency of terrorist attacks. The highest point in terms of the number of attacks occurred in 2011.
  2. However, since 2014, there has been a decline in the number of terrorist attacks, indicating progress in efforts to mitigate such incidents.
  3. The data also reveals the number of people killed each year due to these attacks.



2. No. of People Killed per Year

```python
# Showing Peoples Kiled Per Year
kill_per_year = terrorism[['Year','Killed']].groupby('Year').sum()
kill_per_year.plot(kind='bar', figsize=(15,5), title = 'No. of People Killed Per Year')
plt.xlabel('Attacking Years')
plt.ylabel('No.of Peoples Killed ')
plt.show()
```

Insights

As time progresses, there has been a significant increase in the number of people harmed or killed due to terrorist activities.

The highest point in terms of the number of people harmed was observed in 2007.

3.  No. of People Wounded Per Year

```python
wounded_per_year = terrorism[['Year','Wounded']].groupby('Year').sum()
wounded_per_year.plot(kind='bar', figsize=(15,5), title = 'No. of People Wounded Per Year')
plt.xlabel('Attacking Years')
plt.ylabel('No.of Peoples Wounded')
plt.show()
```

Insights

The number of people wounded due to terrorist incidents has shown a substantial increase over time.

The highest point in terms of the number of people harmed was observed in 2001.

Although there has been a decline in the number of people harmed since 2007, the figures still remain alarmingly high, emphasizing the urgent need for further reduction in these numbers.
